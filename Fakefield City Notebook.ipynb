```python
#Initialise the class
field = field(x_holdings = 2, y_holdings = 2, x_blocks = 8, y_blocks = 8, block_length = 8)
```

```python
#Set the initial development seed
field.set_seed()
```

```python
#Plot graph
plot(field.g)
```

```python
# Run the model to set development occuring
n_interations = 100

for i in range (0, n_iterations):

    #Calculate the type of development required based on the mix of existing development
    field.required_development()
    
    #Choose the type of land use require the most
    df = field.required_dev_df
    lu = df[df['prop_diff'] == max(df['prop_diff'])].index[0]
    
    #Make the development
    field.make_application(lu)
    
    #Plot subdivision (matplot lib polygon patches)
    field.plot_field()
```
